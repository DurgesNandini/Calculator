```python
from tkinter import*

#for take numbers and convert into string
def btnclick(n):
    global operator
    operator=operator+ str(n)
    text_input.set(operator)
#for clear the screen of calculator
def btnclear():
    global  operator
    operator=""
    text_input.set(operator)
#for calculate all operators through eval() function    
def btnEql():
    global operator
    s=str(eval(operator))
    text_input.set(s)
    operator=""


root=Tk()
root.title('Calculator')
operator=""
text_input=StringVar()
display=Entry(root,font=('arial',20,'bold'),textvariable=text_input,bd=40,insertwidth=4,bg='pink',justify='right').grid(columnspan=4)
#buttons
button1=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='1',command=lambda:btnclick(1))#lambda is used for anonmas value
button1.grid(row=1,column=0)
button2=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='2',command=lambda:btnclick(2))
button2.grid(row=1,column=1)
button3=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='3',command=lambda:btnclick(3))
button3.grid(row=1,column=2)
buttonAdd=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='+',command=lambda:btnclick("+"))
buttonAdd.grid(row=1,column=3)

button4=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='4',command=lambda:btnclick(4))
button4.grid(row=2,column=0)
button5=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='5',command=lambda:btnclick(5))
button5.grid(row=2,column=1)
button6=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='6',command=lambda:btnclick(6))
button6.grid(row=2,column=2)
buttonSub=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='-',command=lambda:btnclick("-"))
buttonSub.grid(row=2,column=3)

button7=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='7',command=lambda:btnclick(7))
button7.grid(row=3,column=0)
button8=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='8',command=lambda:btnclick(8))
button8.grid(row=3,column=1)
button9=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='9',command=lambda:btnclick(9))
button9.grid(row=3,column=2)
buttonMul=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='*',command=lambda:btnclick("*"))
buttonMul.grid(row=3,column=3)

buttonClear=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='C',command=btnclear)
buttonClear.grid(row=4,column=0)
button0=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='0',command=lambda:btnclick(0))
button0.grid(row=4,column=1)
buttonE=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='=',command=btnEql)
buttonE.grid(row=4,column=2)
buttonDiv=Button(root,padx=16,pady=16,bd=8,fg='black',font=('arial',20,'bold'),text='/',command=lambda:btnclick("/"))
buttonDiv.grid(row=4,column=3)

root.mainloop()
```
